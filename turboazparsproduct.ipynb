```python
import requests as req
from bs4 import BeautifulSoup as bs
import re
a=[]
b=[]

u='https://turbo.az/'
page = req.get(u)
soup = bs(page.content, 'html.parser')
a=[]
b=[]

for x in soup.findAll(attrs={'class':"products-i__price products-i__bottom-text"},limit=5):
  a.append(x.text)
for l in soup.findAll(attrs={'class':'products-i__name products-i__bottom-text'},limit=5):
  b.append(l.text)
k=dict(zip(a,b))
```
