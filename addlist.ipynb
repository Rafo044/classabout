```python
class Sirketler:
  def __init__(self, dosya_adi):
    self.sirket_adlari = ['Apple', 'Google', 'Microsoft', 'Facebook', 'Amazon', 'Alibaba', 'Tencent', 'Intel', 'IBM', 'Cisco', 'Oracle', 'HP', 'Accenture', 'Adobe', 'SAP']
    
    with open(dosya_adi, 'w') as f:  # write acdim
      for sirket in self.sirket_adlari:  # sirketler listini donguye saldim
        f.write(sirket + '\n')  # sirketin adi axira elave olunur
    
  def sirketler_list(self):
    for sirket in self.sirket_adlari:
      print(sirket)
      
  def sirket_elave(self, sirket_adi):
    self.sirket_adlari.append(sirket_adi)
      
  def sirket_sil(self, sirket_ismi):
    self.sirket_adlari.remove(sirket_adi)
```

```python
a=Sirketler("sirketler.txt")
```
