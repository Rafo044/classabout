```python
import numpy as np
import pandas as pd
import re
from bs4 import BeautifulSoup as bs
import requests as req
```

```python
u='https://www.trendyol.com/kagit-urunleri-x-c104011'
a=req.get(u)
print(a)
```

```python
p = bs(a.content, 'html.parser')
l=[]
for x in p.findAll(attrs={'class':"prdct-desc-cntnr-name"},limit = 10):
  b=x.text
  l.append(b)
```

```python
p = bs(a.content, 'html.parser')
k=[]
for y in p.findAll(attrs={'class':"prc-box-dscntd"},limit = 10):
  t=re.sub(' TL','',y.text)
  z=re.sub(',','.',t)
  z=float(z)
  k.append(z)
```

```python
f=dict(zip(l,k))
```

```python
print(f)
```

```python
dS=pd.Series(f)
```

```python
df = dS.to_frame('Qiymet')
```
