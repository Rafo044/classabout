```python
akk={'Akom 75 Top': 75,
 'Akom_100': 100,
 'AKOM_60': 60,
 'İNCİ AGM 70AH': 70,
 'VARTA 100 AH H3 R+': 103}
```

```python
def qaime():
  item = input('Almaq istediyiniz mehsulu qeyd edin: ')
  nov = int(input('nov sayi:'))
  while nov < 5:
    for n, p in akk.items():
      if n == item:    
        quantity = int(input('Mehsul sayini qeyd edin: '))
        sum = p*quantity
        print(f"{item} {quantity} * {p} = {sum}")
        break
qaime()
```
